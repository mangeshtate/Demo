```python
import pandas as pd

data = {
    "Project": ["Solar Farm A", "Wind Turbine X", "Hydropower Y", "Solar Roof Z", "Geothermal Plant P"],
    "Technology": ["Solar", "Wind", "Hydropower", "Solar", "Geothermal"],
    "Capacity (MW)": [150, 300, 200, 50, 100],  # Megawatts
    "Cost (Million $)": [200, 400, 350, 100, 250],  # Project cost
    "Location": ["California", "Texas", "Washington", "Nevada", "Idaho"],
    "Completion Year": [2023, 2024, 2022, 2025, 2023]
}

projects_df = pd.DataFrame(data)

grouped_data = projects_df.groupby("Technology")["Capacity (MW)"].sum()

print("\nTotal Capacity by Technology:")
print(grouped_data)
```

```python
import numpy as np

# Data
projects = ["Solar Farm A", "Wind Turbine X", "Hydropower Y", "Solar Roof Z", "Geothermal Plant P"]
technology = np.array(["Solar", "Wind", "Hydropower", "Solar", "Geothermal"])
capacity = np.array([150, 300, 200, 50, 100])  # MW

# Grouping by Technology
unique_tech, inverse_idx = np.unique(technology, return_inverse=True)
total_capacity = np.zeros(len(unique_tech), dtype=int)

for i, cap in enumerate(capacity):
    total_capacity[inverse_idx[i]] += cap

# Display results
print("\nTotal Capacity by Technology (NumPy):")
for tech, cap in zip(unique_tech, total_capacity):
    print(f"{tech}: {cap} MW")

```

```python
#Standardization
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Your dataset
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
from sklearn.impute import SimpleImputer
scaler=StandardScaler()
energy_df = pd.DataFrame(data) # Re-create the dataframe to work with original values before imputation and scaling
imputer = SimpleImputer(strategy='mean') # Define the imputer
energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = imputer.fit_transform(
    energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]]
)

energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = scaler.fit_transform(
    energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]]
)
print("Data After Standardization")
print(energy_df)
```
